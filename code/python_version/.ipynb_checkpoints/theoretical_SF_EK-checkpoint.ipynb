```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
n = 100
nj = np.arange(1, 10*n+1)
wj = 1/n
r_j = (nj - 0.5)/n
```

```python
r_j = np.logspace(-2, 2, 101)
k_i = np.logspace(-2, 2, 101)
k_i = 2*np.pi*k_i
```

```python
plt.plot(r_j,'.')
plt.yscale('log')
```

```python
plt.plot(k_i,'.')
plt.yscale('log')
```

```python
import scipy.special as sp
```

```python
dk = np.diff(k_i)
```

```python
dk = np.append(dk[0], dk)
```

```python
def bessel_kernel(k,r): 
    return (1. - sp.jv(0,k*r))
```

```python
ni = len(k_i)
nj = len(r_j)
print(nj)
#k_i = k_j# 2*np.pi/r_j
```

```python
K = np.zeros((ni, nj))

for i in range(ni):
    for j in range(nj):
        K[i,j] = 2*dk[i]*bessel_kernel(k_i[i], r_j[j])
```

```python
Kapprox = np.zeros((ni, nj))

Kapprox1 = np.zeros((ni, nj))
Kapprox2 = np.zeros((ni, nj))

for i in range(ni):
    for j in range(nj):
        
        if k_i[i]< 2/r_j[j]:
            Kapprox[i,j] = (k_i[i]**2) *(r_j[j]**2)/2 * dk[i]
            Kapprox1[i,j] = (k_i[i]**2) *(r_j[j]**2)/2 * dk[i]
            Kapprox2[i,j] = 0.
        else:
            Kapprox[i,j] = 2*dk[i]
            Kapprox1[i,j] = 0.
            Kapprox2[i,j] = 2*dk[i]
```

```python
f_3 = 1*k_i/k_i
f_3[k_i>1] = 1e0*k_i[k_i>1]**-3

f_2 = 1*k_i/k_i
f_2[k_i>1] = 1e0*k_i[k_i>1]**-2

f_53 = 1*k_i/k_i
f_53[k_i>1] = 1e0*k_i[k_i>1]**-(5/3)
```

```python
# Moved origin
f_3 = 1*k_i/k_i
f_3[k_i>0.2] = 0.009*k_i[k_i>0.2]**-3

f_2 = 1*k_i/k_i
f_2[k_i>0.2] = 0.045*k_i[k_i>0.2]**-2

f_53 = 1*k_i/k_i
f_53[k_i>0.2] = 0.075*k_i[k_i>0.2]**-(5/3)
```

```python
# normalize to have same total energy
f_3 = f_3/np.sum(f_3*dk)
f_2 = f_2/np.sum(f_2*dk)
f_53 = f_53/np.sum(f_53*dk)
```

```python
plt.figure(figsize=(4,3))
plt.plot(k_i, f_53, '-', label='-5/3')
plt.plot(k_i, f_2, '-', label = '-2')
plt.plot(k_i, f_3, '-', label='-3')

#plt.ylim([0.1, 10])
plt.xscale('log')
plt.yscale('log')
plt.grid()

plt.legend()

plt.xlabel('k')
plt.ylabel(r'$\hat{E} (k)$')

plt.tight_layout()
plt.savefig('ek_ideal.pdf')
```

### Regularization 

```python
I = np.eye(ni) # get's appended below A

# this weighting is done because we want are thinking of integrals instead of sums 
for i in range(ni):
    I[i, i] = dk[i] # weight by the dk 
    
# Tikhonov regularization but with gradients 
L = np.eye(ni) # get's appended below A


for i in range(ni-1):
    L[i, i+1] = -1 #dk_f[i+1]
    
#L[-1,:] = 0 #If we don't want to use epsilon_u for regularization    

O = np.zeros((ni,)) # gets appended below y
```

```python
lam = 5
mu = 1

# Append the extra matrices, and then solve as simple least squares
K_tik = np.concatenate((K.T, lam*L, mu*I)) 
Kapprox_tik = np.concatenate((Kapprox.T, lam*L, mu*I)) 
```

```python
# Scale independent noise
alpha = 1e-3
y_3 = np.concatenate((g_3+eta_j*alpha, O,O))
f_3inv_tik = np.dot(np.linalg.pinv(K_tik), y_3) 
#f_3inv_approx_err = np.dot(np.linalg.pinv(Kapprox.T), g_3+eta_j*alpha) 

y_2 = np.concatenate((g_2+eta_j*alpha, O,O))
f_2inv_tik = np.dot(np.linalg.pinv(K_tik), y_2) 
#f_2inv_approx_err = np.dot(np.linalg.pinv(Kapprox.T), g_2+eta_j*alpha) 

y_53 = np.concatenate((g_53+eta_j*alpha, O,O))
f_53inv_tik = np.dot(np.linalg.pinv(K_tik), y_53) 
#f_53inv_approx_err = np.dot(np.linalg.pinv(Kapprox.T), g_53+eta_j*alpha) 
```

```python
# Scale dependent noise
alpha = 0.1
y_3 = np.concatenate((g_3+g_3*eta_j*alpha, O,O))
f_3inv_tik = np.dot(np.linalg.pinv(K_tik), y_3) 
#f_3inv_approx_err = np.dot(np.linalg.pinv(Kapprox.T), g_3+eta_j*alpha) 

y_2 = np.concatenate((g_2+g_2*eta_j*alpha, O,O))
f_2inv_tik = np.dot(np.linalg.pinv(K_tik), y_2) 
#f_2inv_approx_err = np.dot(np.linalg.pinv(Kapprox.T), g_2+eta_j*alpha) 

y_53 = np.concatenate((g_53+g_53*eta_j*alpha, O,O))
f_53inv_tik = np.dot(np.linalg.pinv(K_tik), y_53) 
#f_53inv_approx_err = np.dot(np.linalg.pinv(Kapprox.T), g_53+eta_j*alpha) 
```

```python

f_3invapprox_tik = np.dot(np.linalg.pinv(Kapprox_tik), y_3) 
#f_3inv_approx_err = np.dot(np.linalg.pinv(Kapprox.T), g_3+eta_j*alpha) 

f_2invapprox_tik = np.dot(np.linalg.pinv(Kapprox_tik), y_2) 
#f_2inv_approx_err = np.dot(np.linalg.pinv(Kapprox.T), g_2+eta_j*alpha) 

f_53invapprox_tik = np.dot(np.linalg.pinv(Kapprox_tik), y_53) 
#f_53inv_approx_err = np.dot(np.linalg.pinv(Kapprox.T), g_53+eta_j*alpha) 
```
